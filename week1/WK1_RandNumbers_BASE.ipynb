```python
# ------------------------------------------------------------
# üéØ Random Number Guesser
# Purpose: Prompt the user to guess a number until they get it right
# ------------------------------------------------------------

# Ask the user for their first guess and convert input (string) to integer
guess = int(input("What is your guess?: "))

# Set the correct answer (you could randomize this later using random.randint)
correct_number = 5

# Keep looping until the user's guess matches the correct number
while guess != correct_number:
    # Ask again if the guess was incorrect
    guess = int(input("Wrong guess. Try again: "))

# Once the condition is no longer true, the loop ends
print("You got the right answer! üéâ")
```

```python
# Random Number Guesser Using While Loop

# Ask the user for their first guess and convert input (string) to integer
guess = int(input("What is your guess?: "))
correct_number = 5
guess_count = 1

while guess != correct_number:  # While loop
    guess = int(input("What is your guess?: "))  # Keep asking until user inputs the correct number
    guess_count += 1

print(f"Congrats! The right answer was {correct_number}. It took you {guess_count} guesses.")
```
