```python
# Boolean and If Statements
light_is_on = True

if light_is_on: 
    print("The light is on!") # code only runs if True
    print("The light is on" + ", and it is purple!") # Boolean and concatenation
```

```python
light_is_on = False

if light_is_on:
    print("The light is on!") # code does not run if False
```

```python
# Boolean and Else Statements
light_is_on = False

if light_is_on:
    print("The light is on!")
else:
    print("We are in the dark")
```

```python
# Variable with If and Else Statements
weight = 146.5

if weight < 126.5:
    print("You're underweight :)")
else:
    print("You're overweight!")
```

```python
# Variable with If and Else Statements
age = 58

if age >= 18:
    print("This is an adult")

else:
    print("This is a child")
```

```python
# Variable, Comparisons with If and Else Statements
number = 7
remainder = number % 2

if remainder == 0:
    print(f"{number} is even")
else:
    print(f"{number} is odd")
```

```python
def is_even(number):
    remainder = number % 2
    print(remainder)

    if remainder == 0:
        return True
    else:
        return False


# Test it
print(is_even(4))  # Should return True
print(is_even(5))  # Should return False
```
