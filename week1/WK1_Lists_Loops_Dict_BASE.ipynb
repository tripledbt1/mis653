```python
# Function to determine the sum of all even numbers in a list


def sum_even_numbers(numbers):
    total = 0
    for number in numbers:
        remainder = number % 2  # divide by 2 and determine even numbers
        if remainder == 0:  # if remainder is zero, then number is even
            total = total + number
    return total  # return is valid inside a function


# Example usage
numbers = [62, 66, 94, 97, 25, 11, 68, 54, 48, 67]  # List of numbers
result = sum_even_numbers(numbers)
print("Sum of even numbers:", result)
```

```python
# Counter iterates through 'tokens' and returns a dictionary-like object
# mapping each unique word to the number of times it appears.

from collections import Counter  # 1Ô∏è‚É£ import the Counter class from the collections module

text = "a b c a a b"             # 2Ô∏è‚É£ your raw text
tokens = text.split()            # 3Ô∏è‚É£ split text into a list of words; use variable "tokens"

word_count = Counter(tokens)     # 4Ô∏è‚É£ count how often each unique token appears
print(word_count)
print(word_count.most_common(3))  # Counter automatically loops through 'tokens' and tallies how many times each unique word appears, like a dictionary of counts.
```

```python
# How many times a particular word appears in text
text = """
a b c a A b
"""

# .lower() standardizes the text ‚Äî ensuring 'A' and 'a' are treated as the same word.

print(text.lower().split())  # .split() produces a list of words
print(len(text.lower().split()))  # print amount of words
```

```python
# This avoids running .lower() twice and makes the code cleaner

tokens = text.lower().split()
print(tokens)
print(len(tokens))
word_count = Counter(tokens)
print(word_count)
print(word_count.most_common(3))
```
