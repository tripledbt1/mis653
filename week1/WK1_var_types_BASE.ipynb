```python
# Numbers: integers and floats - can be positive or negative
day = 21  # integer
temperature = 72.5  # float
weight_in_pounds = 146.5  # float
height_in_inches = 65.5  # float
age = 58  # integer

# Math without variables
print(3 + 4)

# Math with variables
print(day + 3)
```

```python
temperature = 72.5  # float
print(temperature + 3)

print(weight_in_pounds * 0.15)  # 15 percent of weight in pounds

print(weight_in_pounds - (weight_in_pounds * 0.15))  # weight after losing 15 percent
```
