```python
# Define a function with two integers

def has_remainder(number_1, number_2):  # Define function that takes two ints (no operaters in functions)
    return number_1 % number_2 != 0   # Modulo operator (%) gives remainder after division

# Example test values
print(has_remainder(10, 3))  # True (since 10 √∑ 3 leaves remainder 1)
print(has_remainder(8, 4))   # False (since 8 √∑ 4 divides evenly)
```

```python
# ----------------------------------------------------------
# Function: has_remainder()
# Purpose:  Determines whether dividing one integer by another
#           produces a remainder.
# ----------------------------------------------------------

def has_remainder(number_1, number_2):  # Define the function with two integers
    """
    Takes two integers and returns True if the division
    of number_1 by number_2 leaves a remainder, and False otherwise.
    """
    return number_1 % number_2 != 0  # Modulo (%) computes remainder; comparison (!= 0) evaluates to Boolean True if remainder exists, False otherwise.


# ----------------------------------------------------------
# Example Test Cases
# ----------------------------------------------------------
print(has_remainder(10, 3))  # Expected output: True (10 √∑ 3 leaves remainder 1)
print(has_remainder(8, 4))   # Expected output: False (8 √∑ 4 divides evenly)
print(has_remainder(9, 5))   # Example: True (remainder 4)
```

```python
# ------------------------------------------------------------
# üéØ Random Number Guesser
# Purpose: Prompt the user to guess a number until they get it right
# ------------------------------------------------------------

# Ask the user for their first guess and convert input (string) to integer
guess = int(input("What is your guess?: "))

# Set the correct answer (you could randomize this later using random.randint)
correct_number = 5

# Keep looping until the user's guess matches the correct number
while guess != correct_number:
    # Ask again if the guess was incorrect
    guess = int(input("Wrong guess. Try again: "))

# Once the condition is no longer true, the loop ends
print("You got the right answer! üéâ")
```

```python
# Random Number Guesser Using While Loop

# Ask the user for their first guess and convert input (string) to integer
guess = int(input("What is your guess?: "))
correct_number = 5
guess_count = 1

while guess != correct_number:  # While loop
    guess = int(input("What is your guess?: "))  # Keep asking until user inputs the correct number
    guess_count += 1

print(f"Congrats! The right answer was {correct_number}. It took you {guess_count} guesses.")
```
