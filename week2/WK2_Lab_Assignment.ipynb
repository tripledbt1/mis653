```python
# Lab-Assignment 1: 
# Write a program to create 3 variables for 3 test scores.
# Calculate the total and average of the scores. 
# Pint the total and average scores.

# Create three variables for three test scores
test_score_1 = 83
test_score_2 = 94
test_score_3 = 76

# Calculate the total of the scores
total = test_score_1 + test_score_2 + test_score_3

# Calculate the average of the scores
average = total / 3

# Print the total and average scores
print("Total Score:", total)
print("Average Score:", round(average, 2))  # rounded to 2 decimal places
```

```python
# Lab-Assignment 2: Edit Lab-Assignment 1.
# Check if score inputted for each of the three test scores is greater than zero.
# If score > 0, then consider that score when computing the sum and average of the scores.
# If any score <= 0, then do not use that score in the computation of the total and average.
# Print the total and average scores.

# Ask user to enter each of the three test scores and convert to a float
test_score_1 = float(input("Enter Test Score 1: "))  # asks user to type first score
test_score_2 = float(input("Enter Test Score 2: "))  # asks user to type second score
test_score_3 = float(input("Enter Test Score 3: "))  # asks user to type third score

# Counter for valid test scores
count = 0  # starts at 0; increases by 1 for every valid score > 0 entered

# Total of valid scores
total = 0  # starts at 0; adds up each valid score to find the overall sum

# Check if each score is greater than zero before adding it
if test_score_1 > 0:
    total += test_score_1  # add it to the running total (total + test_score_1)
    count += 1  # add 1 to the couter since this score is valid
else:
    print("Score 1 was ignored because it is less than or equal to zero.")  # runs only if test_score_1 <= 0

if test_score_2 > 0:
    total += test_score_2  # add it to the running total (total + test_score_1 + test_score_2)
    count += 1  # add 1 to the counter since this score is valid
else:
    print("Score 2 was ignored because it is less than or equal to zero.")  # runs only if test_score_2 <= 0

if test_score_3 > 0:
    total += test_score_3  # add it to the running total (total + test_score_1 + test_score_2 + test_score_3)
    count += 1  # add 1 to the counter since this score is valid
else:
    print("Score 3 was ignored because it is less than or equal to zero.")  # runs only if test_score_3 <= 0

# Calculate the average only if there are valid scores
if count > 0:
    average = total / count
else:
    average = 0  # Avoid division by zero if no valid scores

# Print the total and average scores
print("\nTotal Score:", total)
print("Average Score:", round(average, 2))  # Rounded to 2 decimal places
```
