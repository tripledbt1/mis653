## Jupyter Basics Quiz

1. Which menu gives you an option to restart your instance of Python? The kernel menu allows you to interact with the Python instance that runs your code.
2. What is the primary difference between Edit Mode and Command Mode? Edit mode is where we write code and markdown, while command mode is where we can navigate our notebook and edit entire cells.
3. What will happen if the notebook server is closed? Closing your notebook server will cause you to disconnect from your Python kernel.
4. How can multiple values be printed with the print function? Multiple values can be printed if you separate the values with commas.
5. What is one of the differences between Google Colab and Jupyter Notebooks? While Jupyter notebooks can be cloud based, Google's Colaboratory service is always run in the cloud.
