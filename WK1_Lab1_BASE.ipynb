```python
# Lab-Assignment 1: Write a program to create 3 variables for 3 test scores, and then calculate the total and average of the scores. Print the total and average scores.

# Create threes variables for three test scores
score_1 = 83
score_2 = 94
score_3 = 76

# Calculate the total of the scores
total = score_1 + score_2 + score_3

# Calculate the average of the scores
average = total / 3

# Display results
print("Test Scores:", score_1, score_2, score_3)
print("Total Score:", total)
print("Average Score:", round(average, 2))  # rounds to 2 decimal places
```
