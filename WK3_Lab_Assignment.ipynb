```python
# Lab-Assignment 3:
# Define a function that takes 2 scores as arguments.
# Compute the average of these 2 scores inside the function.
# The function should return the average value.
# Write a Python program to call this function.
# Once the value of average score is returned to the main program, print this value.

# Define a function that takes two numeric scores
def compute_avg_scores(score1, score2):
    """
    Receives two numeric scores as input,
    computes their average, and returns the result.
    """
    average = (score1 + score2) / 2
    return average

# Ask the user to enter two test scores (floats)
score_1 = float(input("Enter first score: "))   # asks user to type first score
score_2 = float(input("Enter second score: "))  # asks user to type second score

# Call the function and store the returned value
average_score = compute_avg_scores(score_1, score_2)

# Print the result (rounded to 2 decimal places)
print("\nThe average of", score_1, "and", score_2, "is", round(average_score, 2))
```
